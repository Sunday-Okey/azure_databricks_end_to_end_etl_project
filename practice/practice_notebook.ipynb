```python
dbutils.widgets.text("file_name", "")
```

```python
file_name = dbutils.widgets.get("file_name")
print(file_name)
```

```python
df = spark.read.format("parquet")\
    .load(f"abfss://source@storageaccountsunnybest.dfs.core.windows.net/{file_name}")


df.selectExpr(
    "*",
    "_metadata.file_size",
    "_metadata.file_modification_time",
    "_metadata.file_name",
    "_metadata.file_path"
).write.mode("overwrite").saveAsTable(f"databricksdemo.practice.{file_name}")
```
